```python
# print pylarc.hash_from_four_integers(5,22,0,4,30)

h_exp = 15
def random_hash(h_exp):
    h_size = 2**h_exp
    a = [0 for i in range(4)]
    for i in range(4):
        a[i] = np.random.randint(1,h_size)   
    # print a    
    return pylarc.hash_from_four_integers(a[0],a[1],a[2],a[3],h_exp)
                                       
h_table = [0 for i in range(2**h_exp)]  


for j in range(1000000):
    h_table[random_hash(h_exp)] += 1

# c = random_hash(h_exp)
# print c
for k in range(3):
    print h_table[k]

# make histogram of the hash table
n, bins, patches = plt.hist(h_table,range(np.max(h_table)+1), normed=0, facecolor='green', alpha=1)
print n, bins, patches
plt.show()
    
```
